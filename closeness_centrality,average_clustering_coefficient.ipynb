```python
import pandas as pd
import networkx as nx
import random
```

```python
def load_adjacency_list(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)
    # Convert the DataFrame to a list of tuples
    edges = [tuple(x) for x in df.values]
    return edges
```

```python
def create_graph_from_adj_list(edges):
    # Create a NetworkX graph from the adjacency list
    G = nx.Graph()
    G.add_edges_from(edges)
    return G
```

```python
%cd /content
csv_file = "roads-usroads/edges.csv"
edges = load_adjacency_list(csv_file)
G = create_graph_from_adj_list(edges)
```

```python
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
import random

def approximate_closeness_centrality(G, sample_size=5000):
    nodes = random.sample(G.nodes(), sample_size)
    return {node: nx.closeness_centrality(G, node) for node in nodes}

closeness_centrality = approximate_closeness_centrality(G, sample_size=5000)
```

```python
import random
def max_closeness_centrality(G,sample_size=5000):
  nodes = random.sample(G.nodes(), sample_size)
  max_val = -1
  max_val = max(nx.closeness_centrality(G, node) for node in nodes)
  return max_val

print(max_closeness_centrality(G,sample_size=5000))
```
