Load the Data and Construct the Graph

```python
import networkx as nx

#Load the graph from the file
def load_graph(file_path):
    G = nx.DiGraph()  # Directed graph
    with open(file_path, 'r') as f:
        next(f)  # Skip the first line as it contains node/edge counts
        for line in f:
            node1, node2 = map(int, line.strip().split())
            G.add_edge(node1, node2)
    return G

# Load the graph
G = load_graph(file_path)
```

Count the Number of Triangles

```python
def count_triangles(G):
    triangles = nx.triangles(G.to_undirected())  # Convert to undirected for triangle counting
    return sum(triangles.values()) // 3  # Each triangle is counted thrice

# Count triangles
num_triangles = count_triangles(G)
print(f"Number of triangles: {num_triangles}")
```

Count the Number of Weakly Connected Components (WCC)

```python
def count_wcc(G):
    wcc = nx.number_weakly_connected_components(G)
    return wcc

# Count weakly connected components
num_wcc = count_wcc(G)
print(f"Number of weakly connected components: {num_wcc}")
```

Count the Number of Strongly Connected Components (SCC)

```python
def count_scc(G):
    scc = nx.number_strongly_connected_components(G)
    return scc

# Count strongly connected components
num_scc = count_scc(G)
print(f"Number of strongly connected components: {num_scc}")
```
